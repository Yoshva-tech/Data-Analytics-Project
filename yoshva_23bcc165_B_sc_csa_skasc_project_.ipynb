```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs for the datasets
confirmed_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
recovered_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"

# Load the datasets
confirmed = pd.read_csv(confirmed_url)
deaths = pd.read_csv(deaths_url)
recovered = pd.read_csv(recovered_url)

# Display the first few rows of the data
print("Confirmed Cases:")
print(confirmed.head())
print("\nDeaths:")
print(deaths.head())
print("\nRecovered Cases:")
print(recovered.head())

# Aggregating data by country
confirmed_by_country = confirmed.groupby("Country/Region").sum().drop(columns=["Lat", "Long"])
deaths_by_country = deaths.groupby("Country/Region").sum().drop(columns=["Lat", "Long"])
recovered_by_country = recovered.groupby("Country/Region").sum().drop(columns=["Lat", "Long"])

# Example analysis: Total cases for a specific country (e.g., India)
country = "India"
confirmed_country = confirmed_by_country.loc[country].iloc[-1]
deaths_country = deaths_by_country.loc[country].iloc[-1]
recovered_country = recovered_by_country.loc[country].iloc[-1]

print(f"\nCOVID-19 Statistics for {country}:")
print(f"Confirmed Cases: {confirmed_country}")
print(f"Deaths: {deaths_country}")
print(f"Recovered: {recovered_country}")

# Visualization: Plotting the trend for a specific country
plt.figure(figsize=(10, 6))
confirmed_by_country.loc[country].plot(label="Confirmed", color="blue")
deaths_by_country.loc[country].plot(label="Deaths", color="red")
recovered_by_country.loc[country].plot(label="Recovered", color="green")

plt.title(f"COVID-19 Trend in {country}")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Simulated COVID-19 dataset
np.random.seed(42)
dates = pd.date_range(start="2024-01-01", periods=30)
regions = ['Region 1', 'Region 2', 'Region 3', 'Region 4']
data = {
    "Date": np.tile(dates, len(regions)),
    "Region": np.repeat(regions, len(dates)),
    "Confirmed": np.random.randint(1000, 10000, len(dates) * len(regions)),
    "Recovered": np.random.randint(500, 7000, len(dates) * len(regions)),
    "Deaths": np.random.randint(10, 300, len(dates) * len(regions)),
}

# Creating a DataFrame
df = pd.DataFrame(data)

# 1. Line plot: Total confirmed cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Confirmed', hue='Region')
plt.title("COVID-19 Confirmed Cases Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Bar Plot: Total confirmed cases by region
region_totals = df.groupby('Region')['Confirmed'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=region_totals, x='Region', y='Confirmed', palette='Blues')
plt.title("Total Confirmed Cases by Region", fontsize=16)
plt.xlabel("Region")
plt.ylabel("Total Confirmed Cases")
plt.tight_layout()
plt.show()

# 3. Pie chart: Percentage of confirmed, recovered, and death cases by region
region_summary = df.groupby('Region')[['Confirmed', 'Recovered', 'Deaths']].sum()

# Pie chart for 'Region 1' as an example
region_1 = region_summary.loc['Region 1']

plt.figure(figsize=(7, 7))
region_1.plot.pie(autopct='%1.1f%%', labels=region_1.index, startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('COVID-19 Case Distribution in Region 1', fontsize=16)
plt.ylabel("")  # Hide ylabel
plt.tight_layout()
plt.show()

# 4. Box plot: Distribution of confirmed cases across regions
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Region', y='Confirmed', palette='Blues')
plt.title("Distribution of Confirmed Cases Across Regions", fontsize=16)
plt.xlabel("Region")
plt.ylabel("Confirmed Cases")
plt.tight_layout()
plt.show()

# 5. Heatmap: Correlation between Confirmed, Recovered, and Deaths
corr_matrix = df[['Confirmed', 'Recovered', 'Deaths']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Between Confirmed, Recovered, and Deaths", fontsize=16)
plt.tight_layout()
plt.show()
```
